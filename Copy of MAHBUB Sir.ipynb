# DATA PREPROCESSING




DEFINING CSV TO A VARIABLE

```python
train = pd.read_csv("/content/drive/Shareddrives/asefshahriar75/Humayra_project/steele.csv")
train_image = []
# show first 5 values as a sample data
train.head()
```

CONVERT THE IMAGES INTO AN ARRAY

```python
for i in tqdm(range(train.shape[0])):
    img = image.load_img('/content/drive/Shareddrives/asefshahriar75/Humayra_project/Steelproject/'+train['FILENAME'][i],target_size=(200,200,3))
    img = image.img_to_array(img)
    img = img/255
    train_image.append(img)
X = np.array(train_image)
```

DROP THE FILENAME COLUMN FROM CSV

```python
y = np.array(train.drop(['FILENAME'],axis=1))
y.shape
```

SPLITING THE DATASETS BY USING VALIDATION SPLIT(10% SPLIT)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.1)
```

```python
print(y_test)
```

# Proposed cnn
Accuracy= 98%
Validation accuracy = 97%


```python
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(3, 3), activation="relu", input_shape=(200,200,3)))
model.add(MaxPooling2D(pool_size=(3, 3)))
model.add(Dropout(0.2))
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(7, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy',keras_metrics.precision(), keras_metrics.recall(),keras_metrics.f1_score()])
```

```python
hist=model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test), batch_size=64)
model.summary()
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss']) or plt.plot(hist)
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
#fig.savefig("Model loss.png")
plt.show()
```
