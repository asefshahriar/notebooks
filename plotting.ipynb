```python
# Commented out IPython magic to ensure Python compatibility.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import datetime as dt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Input
from sklearn.model_selection import train_test_split
import pprint as pp
```

```python
df = pd.read_csv("/content/drive/MyDrive/train.csv")
df = df[(df["item"]==1) & (df['store']==1)] # Store 1 and item 1 only
df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")
```

```python
plt.plot(df['date'],df['sales'])
plt.title('Sales')
plt.xlabel('Date')
plt.ylabel('Units')
plt.show()
```

```python
df.groupby(df.date.dt.year)['sales'].plot(legend=True)
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'times-new-roman'
```

```python
fig, axs = plt.subplots(nrows=5, ncols=1, figsize=(7, 10))

for i in range(5):
  axs[i].plot(df[df.date.dt.year == (2013+i)]['date'], df[df.date.dt.year == (2013+i)]['sales'])
  # axs[i].set_xlabel('Date')
  # axs[i].set_ylabel('Sales(unit)')

fig.add_subplot(111, frame_on=False)
plt.tick_params(labelcolor="none", bottom=False, left=False)


plt.xlabel('Date')
plt.ylabel('Sales(unit)')
plt.savefig('yearly_sales.png', dpi=600)
plt.show()
```
