```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
file_path = '/content/pareto_data.xlsx'  # Replace this with your file path
df = pd.read_excel(file_path)

# Assuming your Excel sheet has columns named 'Defect Type' and 'Frequency'
defect_counts = df.groupby('Defect Type')['Frequency'].sum().sort_values(ascending=False)
defect_percentage = (defect_counts / defect_counts.sum()) * 100
cumulative_percentage = defect_percentage.cumsum()

# Create the Pareto chart
fig, ax1 = plt.subplots()

ax1.bar(defect_counts.index, defect_counts, color='b')
ax1.set_xlabel('Defect Type')
ax1.set_ylabel('Frequency', color='b')

ax2 = ax1.twinx()
ax2.plot(defect_counts.index, cumulative_percentage, color='r', marker='o')
ax2.set_ylabel('Cumulative Percentage', color='r')

plt.title('Pareto Chart of Defect Types')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
file_path = '/content/pareto_data.xlsx'  # Replace this with your file path
df = pd.read_excel(file_path)

# Assuming your Excel sheet has columns named 'Defect Type' and 'Frequency'
defect_counts = df.groupby('Defect Type')['Frequency'].sum().sort_values(ascending=False)
defect_percentage = (defect_counts / defect_counts.sum()) * 100
cumulative_percentage = defect_percentage.cumsum()

# Modify the defect type labels to add line breaks after each word
defect_labels = [label.replace(' ', '\n') for label in defect_counts.index]

# Create the Pareto chart with adjusted labels
fig, ax1 = plt.subplots(figsize=(10, 6))  # Adjust figure size as needed

ax1.bar(defect_labels, defect_counts, color='b')
ax1.set_xlabel('Defect Type')
ax1.set_ylabel('Frequency', color='b')

ax2 = ax1.twinx()
ax2.plot(defect_labels, cumulative_percentage, color='r', marker='o')
ax2.set_ylabel('Cumulative Percentage', color='r')

plt.title('Pareto Chart of Defect Types')
plt.xticks(rotation=0)  # Set rotation to 0 degrees to avoid overlapping
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)  # Adjust bottom margin to provide more space for labels
plt.show()
```
