# Distribution fitter with python

## The easy way
The best-fitted distribution for a set of univariate data can be easily determined using **distfit** package. distfit is a python package for probability density fitting across 89 univariate distributions to non-censored data by residual sum of squares (RSS), and hypothesis testing. Probability density fitting is the fitting of a probability distribution to a series of data concerning the repeated measurement of a variable phenomenon. distfit scores each of the 89 different distributions for the fit with the empirical distribution and return the best scoring distribution.

```python
import numpy as np
try: 
  from distfit import distfit
except:
  !(pip install git+https://github.com/erdogant/distfit/)
  from distfit import distfit
```

```python
# paste or generate your data as array
X = np.array([7.5, 51.250, 58.750, 88.000, 168.750, 156.750, 169.000, 182.250, 222.000, 244.000])
```

image.png

```python
# Initialize model and select all distributions or the popular distributions or anyone specific to your need
dist = distfit(distr='logistic')
#dist = distfit(distr='popular')

# Determine best-fitting probability distribution for data
dist.fit_transform(X)
```

```python
# Print summary of evaluated distributions
print(dist.summary)
```

```python
# Plot results
dist.plot()
```
