```
import numpy as np
from itertools import combinations

def rough_set_classification(data, labels):
  """
  Performs rough set based classification on the given data and labels.

  Args:
    data: A NumPy array of data points.
    labels: A NumPy array of labels.

  Returns:
    A NumPy array of predicted labels.
  """

  # Calculate the lower and upper approximations of each class.
  lower_approximations = []
  upper_approximations = []
  for class_label in np.unique(labels):
    lower_approximation = data[labels == class_label]
    upper_approximation = data
    for attribute in data.columns:
      if attribute != "class":
        lower_approximation = lower_approximation[data[attribute] == class_label]
        upper_approximation = upper_approximation[data[attribute] != class_label]
    lower_approximations.append(lower_approximation)
    upper_approximations.append(upper_approximation)

  # Calculate the decision rules.
  decision_rules = []
  for i in range(len(lower_approximations)):
    for attribute in data.columns:
      if attribute != "class":
        decision_rules.append((attribute, lower_approximations[i] == upper_approximations[i]))

  # Classify the new data points.
  predicted_labels = []
  for data_point in data:
    for decision_rule in decision_rules:
      if decision_rule[1](data_point):
        predicted_labels.append(decision_rule[0])
        break

  return predicted_labels
```
