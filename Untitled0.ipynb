```python
import random
import pandas as pd

# Generate a list of rolls from 1 to 60
rolls = list(range(1, 61))

# Shuffle the list to randomize the order
random.shuffle(rolls)

# Initialize an empty list to store groups of 3
groups = []

# Iterate through the shuffled list and form groups of 3
for i in range(0, len(rolls), 3):
    group = rolls[i:i+3]
    groups.append(group)

# Create a DataFrame from the groups
df = pd.DataFrame(groups, columns=['Person 1', 'Person 2', 'Person 3'])

# Save the DataFrame to an Excel file
df.to_excel('groups_of_3.xlsx', index=False)
```

```python
# prompt: save scaled data in 3 sheets one in 0-1 range, one in 0-10 range, last one in 1-10 range using 3 minmax scaler function

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the data
data = pd.read_excel('/content/dd.xlsx', sheet_name="Dataset")

# Create three MinMaxScaler objects with different ranges
scaler_01 = MinMaxScaler(feature_range=(0, 1))
scaler_010 = MinMaxScaler(feature_range=(0, 10))
scaler_110 = MinMaxScaler(feature_range=(1, 10))

# Fit and transform the data for each scaler
data_scaled_01 = scaler_01.fit_transform(data.drop('Year', axis=1))
data_scaled_010 = scaler_010.fit_transform(data.drop('Year', axis=1))
data_scaled_110 = scaler_110.fit_transform(data.drop('Year', axis=1))

# Create a new Excel file with three sheets
with pd.ExcelWriter('scaled_data.xlsx') as writer:
    # Create a DataFrame with the scaled data for each range
    df_scaled_01 = pd.DataFrame(data_scaled_01, columns=data.columns.difference(['Year']))
    df_scaled_010 = pd.DataFrame(data_scaled_010, columns=data.columns.difference(['Year']))
    df_scaled_110 = pd.DataFrame(data_scaled_110, columns=data.columns.difference(['Year']))

    # Add a 'Year' column to each DataFrame
    df_scaled_01['Year'] = data['Year']
    df_scaled_010['Year'] = data['Year']
    df_scaled_110['Year'] = data['Year']

    # Write each DataFrame to a separate sheet in the Excel file
    df_scaled_01.to_excel(writer, sheet_name='0-1')
    df_scaled_010.to_excel(writer, sheet_name='0-10')
    df_scaled_110.to_excel(writer, sheet_name='1-10')

```
