```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

H = 0.00125E-6
mu_beta = 8.4823E-15
sigma_beta = 6.0016E-15
lamb = 3E-4

time = np.linspace(0, 2E5, 50)

def R(t):
    z = (H - mu_beta * t)/(sigma_beta * t)
    return norm.cdf(z)  * np.exp(-lamb*t)    

reliability = np.array([R(t) for t in time])

plt.plot(time, reliability)
plt.show()
```
