```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# Data
time_intervals = ['t+1', 't+2', 't+1', 't+2', 't+1', 't+2', 't+1', 't+2']
situations = ['Excellent', 'Excellent', 'Good', 'Good', 'Average', 'Average', 'Poor', 'Poor']
calculated_service_level = [0.43, 0.40, 0.23, 0.23, 0.18, 0.20, 0.16, 0.17]

# Create a dictionary to map time intervals to colors
time_interval_colors = {'t+1': 'r', 't+2': 'b'}

# Create a figure and axis
fig, ax = plt.subplots()

# Create x-axis positions for bars
x_pos = np.arange(len(situations))

# Adjust the x-axis positions for 't+1' and 't+2' bars with a small gap
x_adjusted = [x - 0.3 if time_intervals[i] == 't+1' else x - 0.7 for i, x in enumerate(x_pos)]

# Set the desired column width
column_width = 0.6  # Adjust this value as needed

# Create bars for each situation with colors based on time intervals
for i in range(len(situations)):
    time_interval = time_intervals[i]
    color = time_interval_colors[time_interval]
    ax.bar(x_adjusted[i], calculated_service_level[i], color=color, label=time_interval, width=column_width)
    ax.text(x_adjusted[i] + column_width / 2, calculated_service_level[i] + 0.01, f'{calculated_service_level[i]:.2f}', ha='right')

# Set x-axis labels once for each situation
unique_situations = list(set(situations))
unique_situation_indices = [situations.index(situation) for situation in unique_situations]
ax.set_xticks(np.array(unique_situation_indices))
ax.set_xticklabels(unique_situations, rotation=0, fontsize=8)

# Set y-axis label
ax.set_ylabel('Calculated Service Level')
ax.set_xlabel('Working Level Situations')

# Set title
ax.set_title("Organization's Work Level During Malware Attack \n at Time Interval t+1 and t+2")

# Adjust the y-axis limits to add padding at the top
plt.ylim(0, max(calculated_service_level) + 0.1)  # You can adjust the padding as needed

# Create custom legend handles and labels
legend_handles = [mpatches.Patch(color=color, label=time_interval) for time_interval, color in time_interval_colors.items()]

# Add a legend with custom handles
ax.legend(handles=legend_handles, title='Time Interval', loc='upper right')

# Show the plot
plt.show()
```

#Figure 12

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# Data
time_frames = ['No Cyber \nAttack', 'Malware \nAttack', 'Malware \nAttack', 'DDoS \nAttack', 'DDoS \nAttack']
sub_time_frames = ['NA', 't+1', 't+2', 't+1', 't+2']
total_expected_utility = [2617.6, -725.7, -2457.05, -3671, -6630.81]

# Define colors for sub-time frames, including a different color for 'NA'
colors = ['gray' if sub_time == 'NA' else 'b' if sub_time == 't+2' else 'r' for sub_time in sub_time_frames]

# Create a figure and axis
fig, ax = plt.subplots()

# Create x-axis positions for bars
x_pos = np.arange(len(time_frames))

# Create bars for each data point with colors
bars = ax.bar(x_pos, total_expected_utility, color=colors)

# Adjust the y-limits for padding
padding = 1000
ax.set_ylim([min(total_expected_utility) - padding, max(total_expected_utility) + padding])

# Add labels to the bars just above the bars
for i, bar in enumerate(bars):
    text_y = total_expected_utility[i] - (padding / 3.5) if total_expected_utility[i] < 0 else total_expected_utility[i] + (padding / 3.5)
    ax.text(bar.get_x() + bar.get_width() / 2, text_y, str(total_expected_utility[i]), ha='center', va='center')

# Set x-axis labels and rotate if needed
ax.set_xticks(x_pos)
ax.set_xticklabels(time_frames, rotation=0, ha='center')

# Set y-axis label
ax.set_ylabel('Total Expected Utility')

# Set title
ax.set_title('Total Expected Utility of An Organization \n During Different Attacks at Different Time Intervals')

# Create custom legend handles and labels, including 'NA'
legend_handles = [
    mpatches.Patch(color='gray', label='NA'),
    mpatches.Patch(color='r', label='t+1'),
    mpatches.Patch(color='b', label='t+2')
]

# Add a legend with custom handles
ax.legend(handles=legend_handles, title='Sub Time Frames', loc='upper right')

# Add a dotted line along y=0
ax.axhline(0, color='gray', linestyle='--', linewidth=0.8)

# Show the plot
plt.show()
```
